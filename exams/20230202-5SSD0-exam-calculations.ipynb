> (3a)

Since the tosses are independent for a given coin, we get

$$p(D|\mu) = \prod_{k} p(x_k|\mu) = \mu^3 (1-\mu)^2$$

> (3b)  
Since the Beta distribution is conjugate with the Bernoulli distribution, we know that the posterior will be a Beta distribution. We only need to determine the parameters $\alpha$ and $\beta$ for the posterior Beta distribution:
$$\begin{align*}
p(\mu|D) &\propto p(D|\mu) p(\mu) \\
&= \mu^3 (1-\mu)^2 \cdot\mathrm{Beta}(\mu|3,2) \\
&\propto \mu^3 (1-\mu)^2 \cdot \underbrace{\mu^2 (1-\mu)^1}_{\text{from }\mathrm{Beta}(\mu|3,2)} \\
&= \mu^5 (1-\mu)^3 \\
&\propto \mathrm{Beta}(\mu|6,4)
\end{align*}$$

> (3c)
$$\begin{align*}
p(D) &= \int p(D|\mu) p(\mu) \mathrm{d}\mu \\
&= \int \mu^3 (1-\mu)^2 \cdot \frac{\Gamma(5)}{\Gamma(3)\Gamma(2)} \mu^2 (1-\mu)^1\mathrm{d}\mu \\
&= \frac{\Gamma(5)}{\Gamma(3)\Gamma(2)} \int \mu^5 (1-\mu)^3 \mathrm{d}\mu \\
&= \frac{\Gamma(5)}{\Gamma(3)\Gamma(2)} \frac{\Gamma(6)\Gamma(4)}{\Gamma(10)} \int \frac{\Gamma(10)}{\Gamma(6)\Gamma(4)}  \mu^5 (1-\mu)^3 \mathrm{d}\mu \\
&= \frac{\Gamma(4)\Gamma(5)\Gamma(6)}{\Gamma(2)\Gamma(3)\Gamma(10)} \underbrace{\int \mathrm{Beta}(\mu|6,4)\mathrm{d}\mu}_{=1}
\end{align*}$$

> (4d)

The last draw has already been determined, so you have effectively a bag with 5 red balls and 6 green balls. Hence 
$$p(X_1=\mathrm{Red}|X_{12}=\mathrm{Green})=5/11\,.$$ 

Alternatively, if you are a hardcore Bayesian, you don't need to be clever and just get to the answer through sum and product rules:
$$\begin{align*}
p(X_1&=\mathrm{R}|X_{12}= \mathrm{G}) = \frac{p(X_1=\mathrm{R},X_{12}= \mathrm{G})}{p(X_{12}= \mathrm{G})} \\
&= \frac{ \sum_{X_2 \in \{\mathrm{R},\mathrm{G}\}} \dots \sum_{X_{11} \in \{\mathrm{R},\mathrm{G}\}}p(X_1=\mathrm{R},X_2,\ldots,X_{11}, X_{12} = \mathrm{G})}{\sum_{X_1 \in \{\mathrm{R},\mathrm{G}\}}\sum_{X_2 \in \{\mathrm{R},\mathrm{G}\}} \dots \sum_{X_{11} \in \{\mathrm{R},\mathrm{G}\}}p(X_1,X_2,\ldots,X_{11}, X_{12} = \mathrm{G})}

This is straightforward to compute but it's a very long sum, so let's be clever here and just reason to the first answer. For more complex problems (like, what is the probability for the 1st toss to be red if the 5th toss is green but only if the 8th toss is red, etc.), the Bayesian approach may be necessary. 

> (4e)

This is a linear system with only Gaussian sources, so $p(x_n)$ will be Gaussian. Hence we only need to compute the mean and variance:
$$\begin{align*}
\mathbb{E}[x_n] &= \mathbb{E}[\Lambda z_n + v_n] \\
&= \Lambda\mathbb{E}[ z_n] +  \mathbb{E}[ v_n] \\
&= \Lambda \cdot 0 + 0 = 0
\end{align*}$$

$$\begin{align*}
\mathbb{V}[x_n] &= \mathbb{E}[(x_n - \mathbb{E}[x_n])(x_n - \mathbb{E}[x_n])^T] \\
&= \mathbb{E}[x_n x_n ^T] \\
&= \mathbb{E}[(\Lambda z_n + v_n)(\Lambda z_n + v_n)^T] \\
&= \Lambda\mathbb{E}[ z_n z_n^T]\Lambda^T + \underbrace{\mathbb{E}[v_nz_n^T]\Lambda^T + \Lambda\mathbb{E}[z_n v_n^T]}_{=0} + \mathbb{E}[v_n v_n^T] \\
&=  \Lambda\Lambda^T + \Psi
\end{align*}$$

Hence, $p(x_n) = \mathcal{N}(x_n|0,\Lambda\Lambda^T + \Psi)$.
